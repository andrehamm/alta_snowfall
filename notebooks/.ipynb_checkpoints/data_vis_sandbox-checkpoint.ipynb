```python
def plot_scatter(df, x_column_name, y_column_name):
    """Plots a scatter plot of two columns in a pandas dataframe"""
    plt.figure(figsize=(10, 5))
    plt.scatter(df[x_column_name], df[y_column_name])
    plt.xlabel(x_column_name)
    plt.ylabel(y_column_name)
    plt.show()
```

```python
def plot_line(df, x_column_name, y_column_name):
    """Plots a line plot of two columns in a pandas dataframe"""
    plt.figure(figsize=(10, 5))
    plt.plot(df[x_column_name], df[y_column_name])
    plt.xlabel(x_column_name)
    plt.ylabel(y_column_name)
    plt.show()
```

```python
def plot_boxplot(df):
    """Plots a boxplot of a column in a pandas dataframe"""
    fig = plt.figure(figsize=(10, 5))
    nov = df['Nov.'].dropna().to_numpy()
    dec = df['Dec.'].dropna().to_numpy()
    jan = df['Jan.'].dropna().to_numpy()
    feb = df['Feb.'].dropna().to_numpy()
    mar = df['Mar.'].dropna().to_numpy()
    apr = df['Apr.'].dropna().to_numpy()
    data = [nov, dec, jan, feb, mar, apr]
    ax = fig.add_axes([0,0,1,1])
    ax.set_xticklabels(['Nov.', 'Dec.', 'Jan.', 'Feb.', 'Mar.', 'Apr.'])
    bp = ax.boxplot(data)
#     plt.boxplot(df[column_name])
    plt.xlabel("Month")
    plt.ylabel("Snowfall (in.)")
    plt.title("Distribution of snowfall at Alta, UT by month")
    plt.show()
```

```python
def plot_histogram(df, column_name):
    """Plots a histogram of a column in a pandas dataframe"""
    num_bins = 5
    fig, axs = plt.subplots(1,1, figsize = (10, 7))
    N, bins, bars = axs.hist(df[column_name], num_bins)
    grads = ((N**(1/4))/N.max())
    norm = colors.Normalize(grads.min(), grads.max())
    for gn, pn in zip(grads, bars):
        color = plt.cm.viridis(norm(gn))
        pn.set_facecolor(color)
    for i in range(num_bins):
        plt.text(bins[i], N[i], str(int(N[i])))
    plt.title('Snowfall Histogram for Alta, UT')
    plt.xlabel('Snowfall (in.)')
    plt.ylabel("Seasons")
    plt.show()
```

```python
def csv_to_df(csv_path):
    """Reads a csv file and returns a pandas dataframe"""
    df = pd.read_csv(csv_path)
    return df
```

```python
from sklearn.model_selection import train_test_split
x = agg_df['year']
y = agg_df['max_temp']
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
x_train = x_train.values.reshape(-1,1)
x_test = x_test.values.reshape(-1,1)
reg = LinearRegression().fit(x_train,y_train)
yhat = reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print('Mean Squared Error: %.2f' % mean_squared_error(y_test,yhat))
print('RMSE: %.2f' % np.sqrt(mean_squared_error(y_test,yhat)))
print('Mean Absolute Error: %.2f' % mean_absolute_error(y_test,yhat))
print('R2: %.2f' % r2_score(y_test,yhat))
```

```python
from sklearn.linear_model import LinearRegression
x = agg_df[['year', 'max_temp']]
y = agg_df['snowfall']

reg = LinearRegression().fit(x,y)
```

```python
print('Intercept: ', reg.intercept_)
print('Coefficients: ', reg.coef_)
```

```python
year_list = df_f['year'].values.flatten().tolist()
temp_list = df_f['pred_max_temp'].values.flatten().tolist()
snow_list = []
def predicted_snowfall(x1, x2):
    return reg.intercept_ + (reg.coef_[0]* x1) + (reg.coef_[1] * x2)

for year, temp in zip(year_list, temp_list):
    snowfall = predicted_snowfall(year,temp)
    #snowfall = reg.intercept_ +(reg.coef_ [0] * year) + (reg.coef_[1] * temp)
    snow_list.append(snowfall)

df_f['pred_avg_snowfall_in'] = snow_list
df_f.tail()
```

```python
n = len(df_f)
x1 = np.array(year_list)
x2 = np.array(temp_list)
y_fit = predicted_snowfall(x1, x2)
noise = (np.std(agg_df['snowfall'])*2.5) * (np.random.random(n) - 0.5) #I multiplied this by 2.5 because, visually, it seemed to approximate the spread of the data better
y_ran = y_fit + noise
df_f['pred_snowfall_in'] = y_ran
df_f.tail().round(2)
```
